```python
train = pd.read_csv("/kaggle/input/forest-cover-type-prediction/train.csv")
test = pd.read_csv("/kaggle/input/forest-cover-type-prediction/test.csv")
```

```python
display(train,test)
```

```python
pd.options.display.max_columns = 100
```

```python
all_data = pd.concat([train,test])
```

```python
all_data2 = all_data.drop("Cover_Type",1)
```

```python
all_data2["elv_ver"] = all_data2["Elevation"] - all_data2["Vertical_Distance_To_Hydrology"]
all_data2["fire_hydro"] = abs(all_data2['Horizontal_Distance_To_Hydrology']) - all_data2["Horizontal_Distance_To_Fire_Points"]
all_data2["road_hydro"] = abs(all_data2['Horizontal_Distance_To_Hydrology']) - all_data2["Horizontal_Distance_To_Roadways"]
all_data2["road_fire"] = all_data2['Horizontal_Distance_To_Fire_Points'] - all_data2["Horizontal_Distance_To_Roadways"]
```

```python
train2 = all_data2[:len(train)]
test2 = all_data2[len(train):]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid = train_test_split(train2,train["Cover_Type"],test_size=0.25, random_state=0,stratify=train["Cover_Type"])
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=-1, random_state=0)
rf.fit(X_train,y_train)
result = rf.predict_proba(X_valid)
```

```python
from sklearn.metrics import log_loss
log_loss(y_valid,result)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=-1, random_state=0)
rf.fit(train2, train["Cover_Type"])
result2 = rf.predict(test2)
```

```python
sub = pd.read_csv("/kaggle/input/forest-cover-type-prediction/sampleSubmission.csv")
sub["Cover_Type"] = result2
sub.to_csv("type.csv",index=0)
```
