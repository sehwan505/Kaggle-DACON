```python
import pandas as pd
import numpy as numpy
import matplotlib as mpl
import matplotlib.pyplot as  plt

import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv("train.csv")
train.head()
```

```python
test = pd.read_csv("test.csv")
test.head(20)
```

```python
train_fare = train["Fare"]
train_Cabin = train.loc[train["Cabin"] != "NaN"]

for i in train["Sex"]:
    if i == "male":
        train["Sex"] = 1
    else:
        train["Sex"] = 2
for i in test["Sex"]:
    if i == "male":
        test["Sex"] = 1
    else:
        test["Sex"] = 2


category = ["Pclass","Sex"]
for seq in category:
    train[seq].astype("category")
    test[seq].astype("category")
```

```python
fig, axes = plt.subplots(nrows=2)
fig.set_size_inches(18,10)

plt.sca(axes[0])
plt.xticks(rotation=90, ha='right')
axes[0].set(ylabel='Count',title="Cabin")
sns.countplot(data=train_Cabin, x="Cabin", ax=axes[0])

plt.sca(axes[1])
plt.xticks(rotation=90, ha='right')
axes[0].set(ylabel='Count',title="Cabin")
sns.countplot(data=test, x="Cabin", ax=axes[1])
```

```python
from sklearn.ensemble import RandomForestClassifier 
train.isnull().sum()
test.isnull().sum()
```

```python
feature = ["Pclass","Sex","SibSp","Parch"]
test= test.loc[test['Fare'] != "NaN"]
# train["Parch"].fillna(0.38)
# train["SibSp"].fillna(0.5)
# train["Sex"].fillna(1)
# train["Pclass"].fillna(2)
# train["Survived"].fillna(0)    
    
rf_model = RandomForestClassifier()
rf_model.fit(train[feature],train["Survived"])
```

```python
predict = rf_model.predict(X=test[feature])
```

```python
submission = pd.read_csv("gender_submission.csv")
submission

submission["Survived"] = predict

print(submission.shape)
submission.head()
```
